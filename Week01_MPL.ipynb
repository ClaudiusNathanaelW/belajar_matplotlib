```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = [1,2,3,4]
y = [1,4,2,3]

fig, ax = plt.subplots() #Membuat sebuah figure dan axes
ax.plot(x,y) #Melakukan plotting data 
```

```python
plt.plot(x,y)
```

Cara membuat Figure:

```python
fig = plt.figure() #Figure tanpa axes
```

```python
fig, ax = plt.subplots() #Figure dengan axes
```

```python
fig, ax = plt.subplots(2,3) #Figure dengan 2x3 grid axes
```

Object Oriented Style (Membuat Figure dan Axes + Memanggil method dari keduanya)

```python
x = np.linspace(0,2,100) #0 Starting Point, 2 Ending Point, 100 Total data
x
```

```python
fig, ax = plt.subplots() #Membuat sebuah figure dan axes

#plotting tiga variant data pada axes
ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label="cubic")

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
```

PyPlot Style (Mengandalkan Pyplot untuk membuat figure dan axes)

```python
#plotting tiga variant data pada axes
plt.plot(x, x, label="linear")
plt.plot(x, x**2, label="quadratic")
plt.plot(x, x**3, label="cubic")

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
```
