Plot Lifecycle

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {'Item A': 102372, 'Item B': 49321, 'Item C': 23490, 'Item D': 12345, 'Item E': 6789, 'Item F': 4567, 'Item G': 2345, 'Item H': 1234, 'Item I': 789, 'Item J': 456}
data
```

```python
items = tuple(data.keys())
count = tuple(data.values())
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

plt.show()
```

```python
plt.style.available
plt.style.use('bmh')
fig, ax = plt.subplots()
ax.barh(items, count)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)
labels = ax.get_xticklabels()

plt.setp(labels, rotation=45, horizontalalignment='right')

plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
def ribuan(x, pos):
    return f'{int(x/1000)}K'

fig, ax = plt.subplots()
ax.barh(items, count)
formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(title='Contoh Plot', xlabel = 'Jumlah', ylabel = 'Items')
plt.show()

```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(title='Contoh Plot', xlabel = 'Jumlah', ylabel = 'Items')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count), ls='--', color='r')

ax.set(title='Contoh Plot', xlabel = 'Jumlah', ylabel = 'Items')
plt.show()
```

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count), ls='--', color='r')
ax.set(title='Contoh Plot', xlabel = 'Jumlah', ylabel = 'Items')

fig.savefig('sales.png', transparent=False, dpi = 80)
```
